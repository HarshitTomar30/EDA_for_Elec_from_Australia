```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
import os 
import pandas as pd 
```

# Dropping a column

```python
# removing the date column by using drop
df.drop(['date'],axis=1)
```

# Info of dataset

```python
# getting information about the type of data in given dataset
df.info()
```

```python
# Checking for empty value in any cell
df.isnull().sum()
```

```python
numeric_columns = ['nswprice', 'nswdemand', 'vicprice', 'vicdemand', 'transfer']
statistics = df[numeric_columns].describe()
print(statistics)
```

# Use of groupby 

```python
# counting entry, group by days
df.groupby(by='day').count()
```

# Plotting 


```python
# bar plot
plt.title('Class Distribution')
df['class'].value_counts().plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('Count')

plt.show()
```

```python
# bar chart demand for NSW  and day
sns.barplot(data=df, x='day', y='nswdemand', hue='class')
plt.xlabel('Day')
plt.ylabel('Demand')
plt.show()
```

```python
# Histogram of transfer values 
sns.histplot(data=df, x='transfer', bins=5)
plt.xlabel('Transfer Value')
plt.ylabel('Count')
plt.show()
```

```python
# violin plot of demand for NSW by class
plt.title('Price Distri. by Demanmd')
sns.violinplot(data=df, x='class', y='nswdemand')
plt.xlabel('Class')
plt.ylabel('Demand')
plt.show()
```

```python
#Box plot of demand distribution by day and class
sns.boxplot(data=df, x='day', y='nswdemand',hue ='class')
plt.xlabel('Day')
plt.ylabel('Demand')
plt.show()
```

# Plotting pairwise relationship between variables

```python
 # pair plot of numeric
sns.pairplot(data=df[['nswprice', 'nswdemand', 'vicprice', 'vicdemand', 'transfer']])
plt.show()
```

```python
# heatmap of numeric column
corr = df[['nswprice', 'nswdemand', 'vicprice', 'vicdemand', 'transfer']].corr()
sns.heatmap(corr, annot=True, cmap='inferno')
plt.show()
```

# Relationship between two variables measures for same individual or column

```python
# scatter plot of price and period
sns.scatterplot(data=df, x='period', y='nswprice', hue='period', palette='cool', size='vicprice')
plt.xlabel('period')
plt.ylabel('Price')
plt.title('Period vs Price')
plt.show()
```

```python
# scatter plot of demand and class
sns.scatterplot(data=df, x='nswdemand', y='class', hue='class', palette='dark', size='day')
plt.xlabel('Demand')
plt.ylabel('class')
plt.title('Demand vs class')
plt.show()
```

```python
# histogram 
sns.histplot(data=df, x='nswprice', hue ='class' ,multiple='stack')
plt.show()
```
